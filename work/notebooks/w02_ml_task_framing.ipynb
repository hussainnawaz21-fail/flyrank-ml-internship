## 1. My lane as an ML task (type)


## Task Type: Ranking

My chosen lane is Content Refresh Prioritization. The goal is to rank content pages based on how urgently they should be refreshed. Instead of simply deciding whether a page needs updating, the model should prioritize pages from highest to lowest refresh priority so that editors know which pages to update first. This is a ranking problem because the output is an ordered list rather than a simple yes/no prediction.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
# Load the dataset
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Show columns related to ranking the content
ranking_columns = [
    "content_id",
    "impressions_90d",
    "clicks_90d",
    "sessions_90d",
    "days_since_last_update",
    "trend_direction",
    "trend_pct"
]

df[ranking_columns].head()
```

## Target or Proxy

## 3. Success metric

*One metric you can defend. What number means 'good'?*

## Success Metric

The success metric is Precision@K because the goal is to recommend the most important content pages for refresh. A good result means that most of the top-ranked pages truly need attention, allowing editors to focus on the highest-priority content first.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
print("Total content pages:", len(df))

print("\nTop 10 pages with the largest negative trend:")
df.nsmallest(10, "trend_pct")[
    ["content_id", "trend_pct", "trend_direction"]
]
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

## Why ML beats a fixed rule

A simple rule such as "refresh content older than 365 days" cannot capture the full picture because content performance depends on multiple signals working together, including trend percentage, impressions, clicks, sessions, engagement, and content age. These patterns are complex and change over time. A machine learning model can learn from these combined signals to produce a better ranking of refresh priority, helping editors make more informed decisions than a single fixed rule.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
rule_based = df[df["days_since_last_update"] > 365]

print("Pages older than 365 days:", len(rule_based))

print("\nSample pages:")
display(
    rule_based[
        [
            "content_id",
            "days_since_last_update",
            "trend_pct",
            "impressions_90d",
            "clicks_90d"
        ]
    ].head()
)
```
